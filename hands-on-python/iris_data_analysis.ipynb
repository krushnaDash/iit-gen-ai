```python
from os import name
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load the iris dataset : https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data
# read the data
cloumn_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris_df = pd.read_csv("Data/iris.data.txt", header=None, names=cloumn_names)
print(iris_df.head())

```

```python
sepal_length=np.array(iris_df['sepal_length'])
sepal_witdth=np.array(iris_df['sepal_width'])
mean_sepal_length=np.mean(sepal_length)
std_sepal_width=np.std(sepal_witdth)
print("Mean sepal length is ", mean_sepal_length)
print("SD sepal width is ", std_sepal_width)

```

```python
large_sepal=sepal_length > mean_sepal_length
print(large_sepal)
```

```python
print('Flower with large sepal ', sepal_length[large_sepal] )
```

```python
iris_df[iris_df['sepal_length'] > mean_sepal_length]
```

```python
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris_df, hue="class")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
sns.pairplot(iris_df,hue="class")
```

```python
class_map = {
    "Iris-setosa": 0,
    "Iris-versicolor": 1,
    "Iris-virginica": 2
}
iris_df['class'] = iris_df['class'].map(class_map)
print(iris_df.head())
```
